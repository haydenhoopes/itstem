## Round 3: Introduce odds
- Teach students how to calculate odds. 
- I will give a payout odds to the winners. The payout could be potentially high enough to warrant betting regardless of probability, but could be small enough that even a risky bet wouldn't be worth it.

```python
import pandas as pd

uno = pd.read_csv('rook.csv')
```

```python
uno = uno.sample(frac=1, random_state=1).reset_index()
```

### Round 2
Will I draw a red card?
Vote either yes or no.

```python
uno
```

```python
my_points = 0
```

```python
# Calculate probability
red_cards = uno[uno['Number'].isin([1,2,3,5,8,13])]# & (uno['Number'] <= 7)]
num_red_cards = red_cards.shape[0]
num_cards = uno.shape[0]
probability_of_red_card = num_red_cards / num_cards
probability_of_red_card
```

```python
# Probability is less than .5 to get a red card, so I should select False
my_vote = False
```

```python
# Draw the first card
uno.loc[0]
```

```python
my_points += 1
```

Will I draw a colored card? Ie. not black.

```python
# Calculate probability
colored_cards = uno[uno['Color'] != 'Black']
num_colored_cards = colored_cards.shape[0]
num_cards = uno.shape[0]
probability_of_colored_card = num_colored_cards / num_cards
probability_of_colored_card
```

```python
# Probability is greater than 0.5, so I should pick True
my_vote = True
```

```python
uno.loc[1]
```

```python
# I won!
my_points += 1
```

#### Round 3
Introduce odds

Will I draw a green card? If you choose a green card and win you get x4 points.

```python
colored_cards = uno[uno['Color'] == 'Green']
num_colored_cards = colored_cards.shape[0]
num_cards = uno.shape[0]
probability_of_colored_card = num_colored_cards / num_cards
probability_of_colored_card
```

```python
# Estimated winnings if I pick not green
1-probability_of_colored_card
```

```python
# Estimated winnings if I pick green
probability_of_colored_card*4
```
