# Guess the Number!

In this game, students will predict whether or not a card with certain attributes will be picked randomly from a deck. They will submit their results via Google Forms and the results will be analyzed as a class.

## Match 1
1. The next card drawn will be red.
2. The next card drawn will be an even number.
3. The next card will be less than or equal to 5.
4. The next card will be a yellow card.
5. The next card will be between 8 and 13.
6. The next card will be less than 3 and will also be yellow.
7. The next card will be a green 7.
8. The next card will be either green or black.
9. The next card will be either a red card or will be greater than 10.
10. The next card will be both less than 10 and green.
